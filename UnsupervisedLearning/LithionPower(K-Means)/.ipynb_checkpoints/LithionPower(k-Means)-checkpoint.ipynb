```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize']=(12,6)
```

```python
df=pd.read_csv("Dataset/driver-data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
```

```python
df_analyze=df.drop(['id'],axis=1)
kmeans.fit(df_analyze)
```

```python
kmeans.cluster_centers_
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
print(type(kmeans.labels_))
unique,counts=np.unique(kmeans.labels_,return_counts=True)
print(unique,counts)
print(dict(zip(unique,counts)))
```

```python
df_analyze['cluster']=kmeans.labels_
sns.set_style('whitegrid')
```

```python
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=df_analyze,
          hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
kmeans_4=KMeans(n_clusters=4)
df_analyze_4=df.drop(['id'],axis=1)
kmeans_4.fit(df_analyze_4)

print(kmeans_4.cluster_centers_)
print(kmeans_4.labels_)
print(len(kmeans_4.labels_))
print(type(kmeans_4.labels_))
unique,counts=np.unique(kmeans_4.labels_,return_counts=True)
print(unique,counts)
print(dict(zip(unique,counts)))
```

```python
df_analyze_4['cluster']=kmeans_4.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=df_analyze_4,
          hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```
