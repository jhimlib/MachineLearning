<h2> Handle missing values</h2>

```python
#missing Data in any column
df.isna().any()
```

```python
#Mean Imputation:
from sklearn.preprocessing import Imputer

mean_imputer=Imputer(missing_values=np.nan,strategy='mean',axis=1)
mean_imputer =mean_imputer.fit(df.iloc[:,1:12])
imputed_df=mean_imputer.transform(df.iloc[:,1:12].values)

df1=pd.DataFrame(data=imputed_df)
df1
```

```python
#mode Imputation:from sklearn.preprocessing import Imputer

from sklearn.preprocessing import Imputer
cols=df.columns


mean_imputer=Imputer(missing_values=np.nan,strategy='median',axis=1)
mean_imputer =mean_imputer.fit(df.iloc[:,1:12])
imputed_df=mean_imputer.transform(df.iloc[:,1:12].values)


df2=pd.DataFrame(data=imputed_df)
df2
```
